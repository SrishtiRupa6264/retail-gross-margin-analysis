```python
#Converting the Currency in USD 
# 1 Euro (EUR) ≈ 1.1 USD
# 1 British Pound (GBP) ≈ 1.3 USD
# 1 Chinese Yuan (CNY) ≈ 0.13 USD

#Invoice Total USD 
merged_df.loc[(merged_df['Currency']=="EUR"),'Invoice Total USD']=merged_df["Invoice Total"]*(1.1)
merged_df.loc[(merged_df['Currency']=="GBP"),'Invoice Total USD']=merged_df["Invoice Total"]*(1.3)
merged_df.loc[(merged_df['Currency']=="CNY"),'Invoice Total USD']=merged_df["Invoice Total"]*(0.13)

#Line per unit USD
merged_df.loc[(merged_df['Currency']=="EUR"),'Line PerUnit USD']=merged_df["Line Total"]*(1.1)/merged_df["Quantity"]
merged_df.loc[(merged_df['Currency']=="GBP"),'Line PerUnit USD']=merged_df["Line Total"]*(1.3)/merged_df["Quantity"]
merged_df.loc[(merged_df['Currency']=="CNY"),'Line PerUnit USD']=merged_df["Line Total"]*(0.13)/merged_df["Quantity"]
```
